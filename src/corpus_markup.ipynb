```python
# df = pd.DataFrame(columns=['lib', 'qty1', 'qty2'])
df = pd.DataFrame(columns=[
    'filename', 
    'date', 
    'authors', 
    'genres', 
    'names', 
    'topics', 
    'multiple_authors', 
    'leader_in_authors', 
    'is_written_text', 
    'addressed_to_people', 
    'regular_occasion', 
    'has_major_topic']
    )

for i in range(len(filenames)):
    filename = filenames[i][:-4]
    filename = re.sub(r'\-', ' ', filename, flags=re.IGNORECASE)
    filename_split = filename.split('_')
    print(filename_split)

    date = filename_split[0]
    authors = filename_split[1].split('.')
    genres = filename_split[2].split('.')
    names = filename_split[3].split('.')
    topics = []

    multiple_authors = nan #done, need to check
    leader_in_authors = nan #done, manually fix some cases
    is_written_text = nan # done, TODO: fix manually
    addressed_to_people = nan 
    regular_occasion = nan
    has_major_topic = nan

    if len(filename_split) > 4:
        topics = filename_split[4].split('.')
        has_major_topic = True
    else:
        topics = names 
        has_major_topic = False 
    
    if len(authors) > 1:
        multiple_authors = True
    
    leader_in_authors = False
    for aut in [
        'николай ii',
        'ленин',
        'сталин',
        'хрущев',
        'брежнев',
        'горбачев',
        'ельцин',
        'путин',
        'медведев'
    ]:
        if aut in authors:
            leader_in_authors = True



    for aut in [
        'манифест',
        'указ',
        'декларация',
        'телеграмма',
        'постановление',
        'декрет',
        'резолюция',
        'положение',
        'воззвание',
        'инструкция',
        'сообщение',
        'формула',
        'предписание',
        'письмо',
        'статья',
        'газетная-статья',
        'устав',
        'записка',
        'программа',
        'конституция',
        'соглашение',
        'договор',
        'меморандум',
        'заявление'
    ]:
        for genre in genres:
            if aut in genres:
                is_written_text = True

    for aut in [
        'выступление',
        'тв',
        'радио',
        'обсуждение',
        'речь'
    ]:
        for genre in genres:
            if aut in genres:
                is_written_text = False



    addressed_to_people = False
    for aut in [
        'обращение',
        'тв',
        'манифест',
        'радио',
        'блог',
        'воззвание',
        'газетная-статья',
        
    ]:
        for genre in genres:
            if aut in genres:
                addressed_to_people = True
    
    for aut in [
        'письмо',
        'записки',
        
    ]:
        for genre in genres:
            if aut in genres:
                addressed_to_people = False



    

    if leader_in_authors and len(authors) == 1:
        multiple_authors = False

    
    if len(authors) == 1:
        for kw in [
            'кпсс',
            'снк',
            "съезд",
            "вцик",
            "вкпб",
            "ркпб",
            "правительство",
            "цк",
            "партконференция",
            "правда",
            "президиум",
            "ссср",
            "пленум",
            "совет",
            
        ]:
            for author in authors:
                if kw in author:
                    multiple_authors = True

    regular_occasion = False
    for aut in [
        'съезд',
        'совет',
        'конференция',
        
    ]:
        for genre in genres:
            if aut in genres:
                regular_occasion = True
        if len(topics):
            for topic in topics:
                if aut in topic:
                    regular_occasion = True


    # print(authors)

    df.loc[i] = [
    filename, 
    date, 
    authors, 
    genres, 
    names, 
    topics, 
    multiple_authors, 
    leader_in_authors, 
    is_written_text, 
    addressed_to_people, 
    regular_occasion, 
    has_major_topic
    ]

# for file in sorted(os.listdir(directory)):
#     filename = os.fsdecode(file)
#     print(os.path.join(directory, filename))
    # if filename.endswith(".asm") or filename.endswith(".py"): 
    #     # print(os.path.join(directory, filename))
    #     continue
    # else:
    #     continue
```
