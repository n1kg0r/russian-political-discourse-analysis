```python
nlp = spacy.load('ru_core_news_sm')
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
nlp.max_length = 150000000
stopwords_spacy = nlp.Defaults.stop_words
```

```python
lemmas_basefolder = '../lemmatized-20y-window'
period = 'total'
filename = 'lemmas'
full_filename_test = join(lemmas_basefolder, period, filename)
# Path(join(lemmas_basefolder, period2, filename)).is_file()

with open(full_filename_test, 'r') as f:
    text = f.read()
text

# print(text)

finder_threshold_total_ws_2 = BigramCollocationFinder.from_words(text.split(' '))
finder_threshold_total_ws_2.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords_spacy or re.match(r'[^а-я]', w) or '\xad' in w)
finder_threshold_total_ws_2.apply_freq_filter(10)

finder_threshold_total_ws_5 = BigramCollocationFinder.from_words(text.split(' '), window_size=5)
finder_threshold_total_ws_5.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords_spacy or re.match(r'[^а-я]', w) or '\xad' in w)
finder_threshold_total_ws_5.apply_freq_filter(10)

# finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords_spacy or re.match(r'[^а-я]', w) or '\xad' in w or ( 'Surn' or 'Name') in morph.parse(word)[0].tag)
```

```python
finder_threshold_total_ws_2.nbest(bigram_measures.pmi, 100)
```

```python
scored_total_ws_2 = finder_threshold_total_ws_2.score_ngrams(bigram_measures.pmi)
bigrams_sorted_total_ws_2 = sorted([{'bigram': bigram, 'pmi': score} for bigram, score in scored_total_ws_2], key=lambda d: d['bigram'][0], reverse=True)

scored_freq_total_ws_2 = finder_threshold_total_ws_2.score_ngrams(bigram_measures.raw_freq)
scored_freq_dict_total_ws_2 = {bigram: score for bigram, score in scored_freq_total_ws_2}

scored_dice_total_ws_2 = finder_threshold_total_ws_2.score_ngrams(bigram_measures.dice)
scored_dice_dict_total_ws_2 = {bigram: score for bigram, score in scored_dice_total_ws_2}

scored_likelihood_total_ws_2 = finder_threshold_total_ws_2.score_ngrams(bigram_measures.likelihood_ratio)
scored_likelihood_dict_total_ws_2 = {bigram: score for bigram, score in scored_likelihood_total_ws_2}

for bigram in bigrams_sorted_total_ws_2:
    bigram['freq'] = scored_freq_dict_total_ws_2[bigram['bigram']] 
    bigram['likelihood'] = scored_likelihood_dict_total_ws_2[bigram['bigram']] 
    bigram['dice']= scored_dice_dict_total_ws_2[bigram['bigram']]
    bigram['log_dice'] = 14 + log2(bigram['dice'])

```
