```python
def if_elements_in_column_then_feature_is(column, list_of_elements=[], feature_is=True, otherwise=False):
    for column_elem in column:
        for list_elem in list_of_elements:
            if list_elem in column_elem:
                return feature_is 
    return otherwise
```
