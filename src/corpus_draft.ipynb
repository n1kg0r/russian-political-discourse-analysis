```python
def keywords_in_value(x, keywords: list):
    for k in keywords:
        if k in str(x):
            return True 
    return False
```

```python
df_topic_politics = df[
    df['topic'].apply(lambda x: keywords_in_value(x, [
        'политика',
        'общественная жизнь',
        'управление',
        'администрация',
        'общество',
        'армия',
        'религия'
    ]))
]
df_topic_politics
```

```python
df_topic_politics['created']
```

```python
def date_to_year(val):
    if '.' in str(val):
        return str(val).split('.')[-1]
    else:
        return val
```

```python
df_topic_politics['created_year_upd'] = df_topic_politics['created'].apply(date_to_year)
```

```python
df_topic_politics['publ_year_upd'] = df_topic_politics['publ_year'].apply(date_to_year)
```

```python
df_topic_politics[
    df_topic_politics['created_year_upd'] >= '1900'
][
    df_topic_politics['created_year_upd'] <= '2000'
].groupby('created_year_upd').size().plot.line()
```

```python
df_topic_politics[
    df_topic_politics['publ_year_upd'] >= '1900'
][
    df_topic_politics['publ_year_upd'] <= '2000'
].groupby('publ_year_upd').size().plot.line()
```

```python
df_xx_politics = df_topic_politics[
    df_topic_politics['publ_year_upd'] >= '1900'
][
    df_topic_politics['publ_year_upd'] <= '2000'
]
```

```python
df_xx_politics.groupby('publ_year_upd').size()
```
