```python
from dostoevsky.tokenization import RegexTokenizer
from dostoevsky.models import FastTextSocialNetworkModel

tokenizer = RegexTokenizer()
tokens = tokenizer.split('всё очень плохо')  # [('всё', None), ('очень', None), ('плохо', None)]

model = FastTextSocialNetworkModel(tokenizer=tokenizer)

messages = [
    'привет',
    'я люблю тебя!!',
    'малолетние дебилы'
]

results = model.predict(messages, k=2)

for message, sentiment in zip(messages, results):
    # привет -> {'speech': 1.0000100135803223, 'skip': 0.0020607432816177607}
    # люблю тебя!! -> {'positive': 0.9886782765388489, 'skip': 0.005394937004894018}
    # малолетние дебилы -> {'negative': 0.9525841474533081, 'neutral': 0.13661839067935944}]
    print(message, '->', sentiment)
```

```python
def count_sentiments_for_lemma_3gram_contexts(column, data_folder):
    lemma = column
    _3grams = pd.read_csv(data_folder+'/freq_3grams.csv')
    _3grams_containing_lemma = _3grams[_3grams['lemma'].apply(column_contains_lemma, lemma=lemma)]
    _3grams_list = _3grams_containing_lemma['lemma'].to_list()
    sentiments = model.predict(_3grams_list, k=2)
    positives = [s.get('positive') for s in sentiments]
    negatives = [s.get('negative') for s in sentiments]
    _3grams_containing_lemma['sentiments'] = sentiments 
    _3grams_containing_lemma['positive'] = positives
    _3grams_containing_lemma['negative'] = negatives
    _3grams_containing_lemma['positive_weighted'] = _3grams_containing_lemma['positive'] * _3grams_containing_lemma['count_by_period']
    _3grams_containing_lemma['negative_weighted'] = _3grams_containing_lemma['negative'] * _3grams_containing_lemma['count_by_period']

    _3grams_containing_lemma = _3grams_containing_lemma.fillna(0)
    return _3grams_containing_lemma['positive'].sum(), _3grams_containing_lemma['negative'].sum(), _3grams_containing_lemma['positive_weighted'].sum(), _3grams_containing_lemma['negative_weighted'].sum()


    
```
